```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
print(type(model.names), len(model.names))

print(model.names)
```

```python
model.train(data='/content/drive/MyDrive/ssafy_project_1/fish_species/Fish_Data/Fish_Data.yaml', epochs=100, patience=30, batch=32, imgsz=416)
```

```python
print(type(model.names), len(model.names))

print(model.names)
```
